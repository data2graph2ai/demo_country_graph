```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# assuming df_text has these columns with list/array vectors:
# bg_geo_vector, geo_bg_vector, country_bg_geo_vector, country_geo_bg_vector

X_bg_geo         = np.vstack(df_text["bg_geo_vector"].values)
X_geo_bg         = np.vstack(df_text["geo_bg_vector"].values)
X_ctry_bg_geo    = np.vstack(df_text["country_bg_geo_vector"].values)
X_ctry_geo_bg    = np.vstack(df_text["country_geo_bg_vector"].values)

sim_bg_geo      = cosine_similarity(X_bg_geo)
sim_geo_bg      = cosine_similarity(X_geo_bg)
sim_ctry_bg_geo = cosine_similarity(X_ctry_bg_geo)
sim_ctry_geo_bg = cosine_similarity(X_ctry_geo_bg)

n = len(df_text)
countries = df_text["country"].values
iso3s = df_text["iso3"].values
```

```python
pairs = []

for i in range(n):
    for j in range(i + 1, n):
        pairs.append({
            "country1": countries[i],
            "country2": countries[j],
            "code1": iso3s[i],
            "code2": iso3s[j],
            # raw concatenation
            "sim_bg_geo_raw": float(sim_bg_geo[i, j]),
            "sim_geo_bg_raw": float(sim_geo_bg[i, j]),
            # with country template
            "sim_ctry_bg_geo": float(sim_ctry_bg_geo[i, j]),
            "sim_ctry_geo_bg": float(sim_ctry_geo_bg[i, j]),
        })

sim_df = pd.DataFrame(pairs)
sim_df.head(2)
```

```python
import pandas as pd

# 1. Normalize edges and aggregate relationship types per pair
edges_clean = edges.copy()

# clean 'how' a bit (strip, lowercase, remove trailing periods)
edges_clean["how_clean"] = (
    edges_clean["how"]
    .str.strip()
    .str.lower()
    .str.replace(r"\.$", "", regex=True)
)

# create an order-invariant key for the pair
edges_clean["key"] = edges_clean.apply(
    lambda r: tuple(sorted([r["node1"], r["node2"]])),
    axis=1
)

# collect all relationship types per pair
pair_types = (
    edges_clean
    .groupby("key")["how_clean"]
    .agg(lambda x: set(x))
    .reset_index(name="how_set")
)

def classify_relation(how_set):
    has_land = "land" in how_set
    has_sea = "sea" in how_set
    if has_land and has_sea:
        return "both"
    elif has_land:
        return "land"
    elif has_sea:
        return "sea"
    else:
        return "none"

pair_types["border_type"] = pair_types["how_set"].apply(classify_relation)

# mapping: (codeA, codeB) -> "both"/"land"/"sea"/"none"
border_map = dict(zip(pair_types["key"], pair_types["border_type"]))

# 2. Apply this mapping to sim_df

sim_df = sim_df.copy()
sim_df["key"] = sim_df.apply(
    lambda r: tuple(sorted([r["code1"], r["code2"]])),
    axis=1
)

sim_df["border_type"] = sim_df["key"].map(border_map).fillna("none")

# optional: drop helper key
sim_df = sim_df.drop(columns=["key"])

sim_df.tail(2)
```

```python
# Counts per border type
counts = sim_df["border_type"].value_counts().rename("count")

# Percentages per border type
percents = (sim_df["border_type"]
            .value_counts(normalize=True)
            .mul(100)
            .round(2)
            .rename("percent"))

# Combine into one table
border_stats = pd.concat([counts, percents], axis=1)
border_stats

```

```python
def select_pairs_by_sim(
    borders_df,
    col_name,
    n=5,
    which="top",
    border_type="both",
):
    """
    Return top or bottom n pairs by a given similarity column,
    filtered by a single border_type.

    Parameters
    ----------
    borders_df : pd.DataFrame
        DataFrame with columns:
        country1, country2, code1, code2, border_type, and sim columns.
    col_name : str
        Name of the similarity column to sort by (e.g. "sim_geo_bg_raw").
    n : int, optional
        How many rows to return (default 5).
    which : {"top", "bottom"}, optional
        Whether to return the highest ("top") or lowest ("bottom") values.
    border_type : str, optional
        Which border_type to keep, e.g. "land", "sea", "both".
        Default is "both".
    """
    if which not in ("top", "bottom"):
        raise ValueError("which must be 'top' or 'bottom'")

    df = borders_df

    # filter by single border_type string
    if border_type is not None:
        df = df[df["border_type"] == border_type]

    ascending = (which == "bottom")

    return (
        df
        .sort_values(col_name, ascending=ascending)
        [["country1", "country2", "code1", "code2", "border_type", col_name]]
        .head(n)
    )
```

```python
col_name = "sim_geo_bg_raw"
border_types = ["both", "land", "sea"]
n = 5  # top N per border type

frames = []

for bt in border_types:
    df_bt = select_pairs_by_sim(
        borders_sim_df,
        col_name=col_name,
        n=n,
        which="top",
        border_type=bt,
    )
    # drop code1, code2
    df_bt = df_bt.drop(columns=["code1", "code2"])
    frames.append(df_bt)

top_pairs_all = pd.concat(frames, ignore_index=True)

top_pairs_all.head(len(top_pairs_all))
```

```python
col_name = "sim_geo_bg_raw"
border_types = ["both", "land", "sea"]
n = 5  # bottom N per border type

frames_bottom = []

for bt in border_types:
    df_bt = select_pairs_by_sim(
        borders_sim_df,
        col_name=col_name,
        n=n,
        which="bottom",   # <-- key change
        border_type=bt,
    )
    # drop code1, code2
    df_bt = df_bt.drop(columns=["code1", "code2"])
    frames_bottom.append(df_bt)

bottom_pairs_all = pd.concat(frames_bottom, ignore_index=True)

bottom_pairs_all.head(len(bottom_pairs_all))
```
